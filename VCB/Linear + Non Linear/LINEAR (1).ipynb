```python
#Import libraries
import numpy.polynomial.polynomial as poly
from datetime import datetime
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Get the data
df = pd.read_csv("VIC.csv")
df
```

```python
#get infomation data
df.info()
```

```python
#Show description data
df.describe()
```

```python
#Create Time column has value is index
df['Time'] = np.arange(len(df.index))
df
```

```python
#
df['new_time'] = pd.to_datetime(df.Date)
type(df['new_time'])
```

```python
df.index = df['new_time']
df
```

```python
##visualization FPT stock price from 2015 to 2022
df.plot(x = 'Time', y = 'Price', figsize = (12,5))
```

```python

x = np.array(df.Time, dtype = float)
y = np.array(df.Price, dtype = float)
z = [1752,1753,1754,1755,1756,1757,1758,1759,1760,1761,1762,1763,1764,1765,1766,1767,1768,1769,1770,1771,1772,1773,1774,1775,1776,1777,1778,1779,1780,1781]
```

```python
no_of_days = 30
future_day = pd.date_range(start= "12/1/2022",end = "12/30/2022")
```

```python
coefs = poly.polyfit(x, y, 1)
x_new = np.linspace(x[0], x[-1] + no_of_days, num = len(x) + no_of_days)
fit = poly.polyval(x_new, coefs)
pred = poly.polyval(z, coefs)
predictions = pd.DataFrame(future_day ,pred)
print(predictions)
```

```python
from sklearn.metrics import mean_absolute_percentage_error, mean_squared_error

fit = fit[30:].reshape(1752,)

mape = mean_absolute_percentage_error(y, fit)
mse = mean_squared_error(y, fit)
rmse = np.sqrt(mse)


print(f"MAPE: {mape * 100:.2f}%")
print(f"RMSE: {rmse:.2f}")
```
